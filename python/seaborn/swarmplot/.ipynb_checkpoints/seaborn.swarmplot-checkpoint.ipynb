```python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
ax = sns.swarmplot(x=tips["total_bill"])
```

```python
ax = sns.swarmplot(x ="day",
                   y = "total_bill",
                   data = tips)
```

```python
ax = sns.swarmplot(x ="total_bill",
                   y = "day",
                   data = tips)
```

```python
ax = sns.swarmplot(x ="day",
                   y = "tip",
                   data = tips)
```

```python
ax = sns.swarmplot(x="day", 
                   y="total_bill", 
                   hue="sex", 
                   data=tips,
                   dodge = True)
```

```python
ax = sns.swarmplot(x="day", 
                   y="total_bill", 
                   hue="smoker",
                   data=tips, 
                   palette="Set2", 
                   dodge=True)
```

```python
ax = sns.swarmplot(x="time", 
                   y="total_bill", 
                   data=tips,
                   #order=["Dinner", "Lunch"]
                   size = 5
                  )
```
