```python
import seaborn as sns
import numpy as np
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
ax = sns.swarmplot(x=tips["total_bill"])
```

```python
ax = sns.swarmplot(x ="day",
                   y = "total_bill",
                   data = tips)
```

```python
ax = sns.swarmplot(x ="total_bill",
                   y = "day",
                   data = tips)
```

```python
ax = sns.swarmplot(x ="day",
                   y = "tip",
                   data = tips)
```

```python
ax = sns.swarmplot(x="day", 
                   y="total_bill", 
                   hue="sex", 
                   data=tips,
                   dodge = True)
```

```python
ax = sns.swarmplot(x="day", 
                   y="total_bill", 
                   hue="smoker",
                   data=tips, 
                   palette="Set2", 
                   dodge=True)
```

```python
ax = sns.swarmplot(x="time", 
                   y="total_bill", 
                   data=tips,
                   #order=["Dinner", "Lunch"]
                   size = 5
                  )
```

Draw swarms of observations on top of a box plot:

```python
ax = sns.boxplot(x="total_bill", 
                 y="day", 
                 data=tips, 
                 whis=np.inf)

ax = sns.swarmplot(x="total_bill", 
                   y="day", 
                   data=tips, 
                   color=".2")
```

Draw swarms of observations on top of a violin plot:

```python
ax = sns.violinplot(x="day", 
                    y="total_bill", 
                    data=tips, 
                    inner=None)
ax = sns.swarmplot(x="day", 
                   y="total_bill", 
                   data=tips,
                   color="white", 
                   edgecolor="gray")
```
