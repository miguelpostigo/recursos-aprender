### Plots with Seaborn
To start things off, let's just make a 
- *scatterplot*

based on two variables from the data set.
I'll use HP and Attack in this example, but feel free to do something different!

```python
sns.jointplot(x="HP",
              y="Attack",
              data=df,
              kind="scatter",
              color = "Green",
              height = 10,
              ratio = 4,
              space = 0.2,
              dropna = False,
              #{x,y}lim = [100,100]
              marginal_ticks = True,
              #hue = "Type 1"
              
             );
```
