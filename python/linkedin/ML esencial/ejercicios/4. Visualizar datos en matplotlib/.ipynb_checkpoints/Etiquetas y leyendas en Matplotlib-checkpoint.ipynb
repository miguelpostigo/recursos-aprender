```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(0) #fijar una semilla para que nos salgan los mismos casos
df = pd.read_csv("C:/Users/miguel.postigo/Desktop/carpeta/programacion/python/linkedin/esencial/base_datos_2008.csv",
                nrows = 100000)
data = np.unique(df.DayOfWeek,return_counts = True)
labs = ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"]
data
```

```python
plt.pie(x = data[1],
       #labels = labs,
       radius = 1.7,
       colors = sns.color_palette("hls",7),
        explode = (0,0,0,0.1,0,0,0.1),
       startangle = 90,
       autopct = "%1.1f%%",
       labeldistance = 1.2)
plt.legend(loc = "upper left",labels = labs)
plt.show()
```

```python
plt = sns.barplot(x = labs, y = data[1])
plt.set(xlabel = "Día de la Semana",ylabel = "Número de vuelos")
```
